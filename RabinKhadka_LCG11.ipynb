 Section 7.1 — Warming Up Exercise

```python
# Given data: (study hours, entertainment hours, sleep hours) for 15 days
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
```

 Task 1: Classify Study Time

```python
# empty lists
low_study = []
moderate_study = []
high_study = []

#iterates over the time_data
for day in time_data:
    study_hours = day[0]

    if study_hours < 3:
        low_study.append(study_hours)
    elif study_hours <= 5:
        moderate_study.append(study_hours)
    else:  # > 5
        high_study.append(study_hours)

# Printing the lists
print("Low study hours less than 3 hours :", low_study)
print("Moderate study hours between 3 and 5 hours :", moderate_study)
print("High study hours more than 5 hours :", high_study)
```

Task 2: Count Days per Category

```python
print("Days with low study time:", len(low_study))
print("Days with moderate study time:", len(moderate_study))
print("Days with high study time:", len(high_study))
```

Task 3. Convert Study Hours to Minutes:

```python
study_minutes = []

for day in time_data:
    study_hours = day[0]
    minutes = study_hours * 60
    study_minutes.append(minutes)

print("Study time in minutes:", study_minutes)
```

Task 4. Analyze Average Time Use:

```python
study_list = []
ent_list = []
sleep_list = []

for day in time_data:
    study_list.append(day[0])
    ent_list.append(day[1])
    sleep_list.append(day[2])

def average(lst):
    total = 0
    for x in lst:
        total += x
    return total / len(lst)

print("Average study hours:   ", average(study_list))
print("Average entertainment: ", average(ent_list))
print("Average sleep hours:   ", average(sleep_list))
```

Task 5. Visualization - Study vs Sleep Pattern:

```python
import matplotlib.pyplot as plt

study_values = [day[0] for day in time_data]
sleep_values = [day[2] for day in time_data]


plt.scatter(study_values, sleep_values, color='purple')
plt.title("Study Hours vs Sleep Hours")
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.grid(True)
plt.show()
```

Task 1 - Sum of Nested Lists:

```python
def sum_nested_list(lst):
    """Adds up all numbers in a list, even if lists are inside other lists."""
    total = 0

    for item in lst:
        if type(item) == list:
            total += sum_nested_list(item)
        else:
            total += item

    return total

nested = [1, [2, [3, 4], 5], 6, [7, 8]]
print("Sum of nested list:", sum_nested_list(nested))
```

Task 2: Generate All Permutations of a String

```python
def generate_permutations(s):
    """Returns a sorted list of all possible rearrangements of the letters in s."""
    if len(s) <= 1:
        return [s]

    result = []
    for i in range(len(s)):
        char = s[i]
        rest = s[:i] + s[i+1:]
        for perm in generate_permutations(rest):
            result.append(char + perm)


    return sorted(list(set(result)))

print("Permutations of 'abc':", generate_permutations("abc"))
print("Permutations of 'aab':", generate_permutations("aab"))
```

Task 3: Directory Size Calculation

```python
def calculate_directory_size(directory):
    """Adds up all file sizes in a folder, including files in subfolders."""
    total = 0

    for name, content in directory.items():
        if type(content) == dict:
            total += calculate_directory_size(content)
        else:
            total += content

    return total


folder = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print("Total size:", calculate_directory_size(folder), "KB")
```

 Task 1: Coin Change (Min Coins)

```python
def min_coins(coins, amount):

    if amount == 0:
        return 0
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for x in range(coin, amount + 1):
            dp[x] = min(dp[x], dp[x - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Test
print(min_coins([1, 2, 5], 11))
print(min_coins([2], 3))
```

Task 2: Longest Common Subsequence (LCS)

```python
def longest_common_subsequence(s1, s2):
    m = len(s1)
    n = len(s2)


    dp = []
    for i in range(m + 1):
        dp.append([0] * (n + 1))


    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[m][n]

print("LCS length of 'abcde' and 'ace':", longest_common_subsequence("abcde", "ace"))
```

Task 3 - 0/1 Knapsack Problem:
1. Write a function knapsack(weights, values, capacity) that:
• Uses DP to determine the maximum value that can be achieved within the given
weight capacity.

2. Test with weights [1, 3, 4, 5], values [1, 4, 5, 7], and capacity 7. The re-
sult should be 9.

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [0] * (capacity + 1)

    for i in range(n):
        for w in range(capacity, weights[i] - 1, -1):
            # Try including item i
            dp[w] = max(dp[w], dp[w - weights[i]] + values[i])

    return dp[capacity]
weights = [1, 3, 4, 5]
values  = [1, 4, 5, 7]
capacity = 7

print("Maximum value:", knapsack(weights, values, capacity))
```
